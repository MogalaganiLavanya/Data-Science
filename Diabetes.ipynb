```python
x=df.iloc[:,:-1]
y= df.iloc[:,-1]
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB = GaussianNB()
```

```python
NB.fit(x_train,y_train)
y_pred = NB.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
nb_scores = cross_val_score(NB, x, y, cv=5)
print("Naive Bayes Cross-validation scores:", nb_scores)
print("Average Naive Bayes accuracy:", nb_scores.mean())
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
```

```python
f"the performance of Naive Bayes model is {acc},classification report is {cr} and confusion matrix is {cm}"
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_pred)
cm = confusion_matrix(y_test,y_pred)
cr = classification_report(y_test,y_pred)
```

```python
rf_scores = cross_val_score(rf, x, y, cv=5)
print("Random Forest Cross-validation scores:", rf_scores)
print("Average Random Forest accuracy:", rf_scores.mean())
```

```python
f"the performance of random forest model is {acc},classification report is {cr} and confusion matrix is {cm}"
```
